# Theoretical Questions

Q.1. What is unsupervised learning in the context of machine learning?

Answer ->

- Unsupervised learning is a type of machine learning where the model is trained on data that has no labels or target values.

- In other words, the algorithm tries to find hidden patterns, structures, or relationships in the input data without any explicit guidance.

Types of unsupervised learning :

  - Clustering
 - Dimensionality Reduction
  - Association Rule Learning

Q.2. How does K-Means clustering algorithm work?

Answer ->

The K-Means clustering algorithm is one of the most popular and simple unsupervised learning techniques used to group data points into K clusters based on their similarity.

**Working of K-Means clustering :**  

Step-1 : Choose K

Step-2 : Initialize Centroids

Step-3 : Assign point to nearest Centroid

Step-4 : Update Centroids

Step-5 : Repeat

Step-6 : Output

Q.3. Explain the concept of a dendrogram in hierarchical clustering.

Answer ->

A dendrogram is a tree-like diagram used to visualize the results of hierarchical clustering.
It shows how individual data points (or clusters) are merged or split step by step, helping you understand the structure and relationships among clusters.

Q.4.  What is the main difference between K-Means and Hierarchical Clustering?

Answer ->

**K-Means Clustering :**

- It uses Partitioning method – directly divides data into K clusters
- It works best for spherical clusters
- It is suitable for large datasets
- It is fast and efficience for large datasets
- It has no inherent visualization (but can use scatter plots)

**Hierarchical Clustering :**

- It uses Hierarchical method – builds a hierarchy/tree of clusters
- It can handle arbitrary shapes
- It is suitable for small to medium datasets
- It is slow and computationally heavy for large
- It has Dendrogram clearly shows cluster hierarchy

Q.6. When would you use Silhouette Score in clustering?

Answer ->

The Silhouette Score is a metric used to evaluate the quality of clustering results, helping you understand how well your data points fit within their assigned clusters.

It’s especially useful when you want to compare different clustering algorithms or choose the optimal number of clusters (K).

**Purpose of Silhouette Score :**

You use the Silhouette Score to measure how well each data point has been clustered:

- It tells you how similar a point is to its own cluster compared to other clusters.

**Use-case of Silhouette Score :**

- Choosing the optimal K in K-Means
- Comparing different clustering algorithms
- Detecting poor cluster assignments
- Validating clustering without ground truth

Q.8. Why is feature scaling important in clustering algorithms like K-Means?

Answer ->

Feature scaling is crucially important in clustering algorithms like K-Means, because these algorithms rely heavily on distance calculations (usually Euclidean distance) to decide how similar or dissimilar data points are.

In K-Means, clustering decisions are based on distances between data points and centroids.
If features are on different scales, those with larger numerical ranges dominate the distance computation, leading to biased clustering results.

Q.9.  How does DBSCAN identify noise points?

Answer ->

DBSCAN groups together data points that are closely packed (dense regions)
and marks points that don’t belong to any dense region as noise (outliers).

**Key Parameters in DBSCAN :**

1. ε (epsilon) - Maximum distance between two points for them to be considered neighbors.

2. minPts (minimum points) - Minimum number of points required to form a dense region (cluster).

**Working of DBSCAN :**

1. Pick a point not yet visited.

2. Find all neighbors within distance ε.

3. If the number of neighbors ≥ minPts,

    → mark it as a core point and form a cluster.

4. If it has fewer than minPts neighbors:
    → it’s not a core point. It might later become a border point if it’s near a core point’s cluster.

5. After clustering, any point that:

    - is not a core point, not a border point, and

    - doesn’t belong to any cluster,

      → is labeled as noise (or outlier).

Q.10. Define inertia in the context of K-Means.

Answer ->

inertia is a key concept in evaluating K-Means clustering performance.
It measures how compact and cohesive the clusters are — in other words, how close the data points are to their assigned cluster centers (centroids).

Inertia (also called Within-Cluster Sum of Squares – WCSS) is the sum of squared distances between each data point and the centroid of the cluster it belongs to.

Q.11. What is the elbow method in K-Means clustering?

Answer ->

The Elbow Method is one of the most widely used techniques to help determine the optimal number of clusters (K) in K-Means clustering.

The Elbow Method helps you find the optimal K by using the inertia (within-cluster variation).

**Process of elbow method :**

1. Run K-Means with different values of K
2. For each K, compute the inertia
3. Plot a graph:

    - X-axis → Number of clusters (K)

    - Y-axis → Inertia (sum of squared distances)
4. The inertia decreases as K increases
5. Look for the point where the rate of decrease sharply changes — this point looks like an “elbow” in the curve

Q.12. Describe the concept of "density" in DBSCAN.

Answer ->

In DBSCAN, density around a data point is defined by two key parameters:

1. ε (epsilon) — the radius of the neighborhood around a point.
It defines how close points should be to each other to be considered neighbors.

2. minPts (minimum points) — the minimum number of points required (including the point itself) to form a dense region.

**Types of Points Based on Density :**

1. Core point - A point that has at least minPts points (including itself) within its ε-radius neighborhood.

2. Border point - A point that is within the ε-radius of a core point but has fewer than minPts points in its own neighborhood.

3. Noise point - A point that is not a core and not reachable from any core point.

Q.14. What does a negative Silhouette Score indicate?

Answer ->

The Silhouette Score measures how well each data point fits within its assigned cluster compared to other clusters.

It tells us how separated and cohesive the clusters are.

**Range of Silhouette Score :**

- +1 : Point is well clustered — far away from other clusters
- 0 : Point lies on or near the boundary between clusters
- -1 : Point is likely misclassified — assigned to the wrong cluster

Q.15. Explain the term "linkage criteria" in hierarchical clustering.

Answer ->

The linkage criterion defines how the distance between two clusters is computed — that is, when you have multiple points in each cluster, how do you decide the distance between clusters?

Types of linkage :      

1. Single linkage - Distance between closest points of the two clusters

2. Complete linkage - Distance between farthest points of the two clusters

3. Centroid linkage - Distance between cluster centroids

4. Ward's method - Merges clusters to minimize variance within clusters

Q.16. Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?

Answer ->

K-Means clustering can perform poorly on data with varying cluster sizes or densities because of the assumptions it makes about how clusters should look.

Here’s why:

- Assumes Equal Cluster Size and Density
- Sensitive to Distance Metric (Euclidean Distance)
- Centroid Location Bias
- Struggles with Non-Spherical Shapes

Q.17. What are the core parameters in DBSCAN, and how do they influence clustering?

Answer ->

**There are core parameters in DBSCAN :**

**1. ε (Epsilon) :**

- Definition :
The maximum distance between two points for one to be considered as in the neighborhood of the other.

- Influence :

    - Determines how close points need to be to be considered part of the same cluster.

    - Small ε → many small clusters, more points labeled as noise.

    - Large ε → merges nearby clusters into larger ones, possibly mixing different clusters together.

**2. min_samples :**

- Definition :
The minimum number of points required within an ε-radius to form a core point (a dense region).

- Influence :

    - Controls how dense a region must be to be considered a cluster.

    - Low min_samples → detects small clusters, but may increase noise sensitivity.

    - High min_samples → requires denser regions, can miss smaller clusters.

Q.18.  How does K-Means++ improve upon standard K-Means initialization?

Answer ->

K-Means++ is an improved initialization method for the K-Means algorithm that helps it find better clusters faster and reduces the risk of poor local minima.

**How K-Means++ Works :**

1. Choose the first centroid randomly from the data points

2. Compute distances
3. Select next centroid probabilistically
4. Repeat Steps 2–3 until k centroids are chosen
5. Run standard K-Means using these improved initial centroids

Q.19. What is agglomerative clustering?

Answer ->

Agglomerative clustering is a type of hierarchical clustering algorithm that builds clusters from the bottom up — by merging smaller clusters into larger ones step by step.

**Steps in Agglomerative Clustering :**

1. Start :
Treat each data point as an individual cluster.

2. Compute distances :
Calculate the pairwise distances between all clusters (using a distance metric like Euclidean distance).

3. Merge :
Combine the two closest clusters into a new cluster.

4. Update distances :
Recalculate distances between the new cluster and all other clusters using a linkage criterion (explained below).

5. Repeat Steps 3–4 until the desired cluster structure is obtained.

Q.20. What makes Silhouette Score a better metric than just inertia for model evaluation?

Answer ->

The Silhouette Score is often considered a better metric than inertia (especially for comparing different clustering models) because it evaluates both cluster cohesion and separation, while inertia only measures compactness.

**Why Silhouette Score is Better :**

- Considers both cohesion and separation
- Helps detect overfitting (too many clusters)
-  Works across different algorithms (K-Means, DBSCAN, etc.)
-  Gives an absolute performance measure — not just a relative one like inertia

Q.21. Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a
scatter plot.

Answer ->>

```python
# Import required libraries
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Step 1: Generate synthetic data with 4 centers
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=42)

# Step 2: Apply K-Means clustering
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# Step 3: Visualize the clusters
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

# Step 4: Plot cluster centers
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X', label='Centroids')

plt.title("K-Means Clustering with 4 Centers")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

Q.29. Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart.

Answer ->>

```python
# Import required libraries
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Step 1: Generate synthetic data
X, y = make_blobs(n_samples=500, centers=4, cluster_std=0.6, random_state=42)

# Step 2: Evaluate Silhouette Scores for k = 2 to 5
silhouette_scores = []
k_values = range(2, 6)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    silhouette_scores.append(score)

# Step 3: Display as a bar chart
plt.figure(figsize=(8, 6))
plt.bar(k_values, silhouette_scores, color='skyblue', edgecolor='black')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Scores for K-Means Clustering (k=2 to 5)')
plt.show()
```

Q.31. Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with
decision boundaries.

Answer ->>

```python
# Import required libraries
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np

# Step 1: Generate synthetic data with overlapping clusters
X, y_true = make_blobs(
    n_samples=400,
    centers=3,
    cluster_std=2.0,  # Higher std → more overlap
    random_state=42
)

# Step 2: Apply K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
labels = kmeans.predict(X)
centers = kmeans.cluster_centers_

# Step 3: Create a mesh grid for decision boundaries
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.05),
                     np.arange(y_min, y_max, 0.05))
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Step 4: Plot decision boundaries and clusters
plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, alpha=0.2, cmap='viridis')
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50)
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, marker='X', label='Centroids')

plt.title("K-Means Clustering with Decision Boundaries (Overlapping Data)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

Q.34.  Load the Breast Cancer dataset and compare inertia values for K = 2 to 6 using K-Means. Show results in a

Q.44.  Load the Iris dataset and plot the Silhouette Coefficient for each sample after KMeans clustering.

Answer ->>

```python
# Import required libraries
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
import matplotlib.pyplot as plt
import numpy as np

# Step 1: Load the Iris dataset
iris = load_iris()
X = iris.data

# Step 2: Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 3: Apply K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X_scaled)

# Step 4: Compute Silhouette Coefficients
silhouette_vals = silhouette_samples(X_scaled, labels)
avg_silhouette = silhouette_score(X_scaled, labels)

# Step 5: Plot Silhouette Coefficient for each sample
y_lower, y_upper = 0, 0
yticks = []
plt.figure(figsize=(8, 6))

for i in range(3):  # For each cluster
    cluster_silhouette_vals = silhouette_vals[labels == i]
    cluster_silhouette_vals.sort()
    y_upper += len(cluster_silhouette_vals)
    plt.barh(range(y_lower, y_upper), cluster_silhouette_vals, height=1.0)
    yticks.append((y_lower + y_upper) / 2)
    y_lower += len(cluster_silhouette_vals)

plt.axvline(avg_silhouette, color="red", linestyle="--", label=f"Average = {avg_silhouette:.2f}")
plt.title("Silhouette Coefficients for Each Sample (Iris Dataset)")
plt.xlabel("Silhouette Coefficient Value")
plt.ylabel("Cluster")
plt.yticks(yticks, [f"Cluster {i}" for i in range(3)])
plt.legend()
plt.show()
```

Q.46.  Load the Wine dataset, apply KMeans, and visualize the cluster assignments in a seaborn pairplot (first 4

Q.47. Generate noisy blobs using make_blobs and use DBSCAN to identify both clusters and noise points. Print the
count.

Answer ->>

```python
# Import required libraries
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
import numpy as np

# Step 1: Generate noisy synthetic data
X, y_true = make_blobs(
    n_samples=400,
    centers=4,
    cluster_std=1.0,
    random_state=42
)

# Add random noise points
rng = np.random.RandomState(42)
noise = rng.uniform(low=-10, high=10, size=(40, 2))  # 40 random noise points
X_noisy = np.vstack([X, noise])

# Step 2: Apply DBSCAN
dbscan = DBSCAN(eps=0.8, min_samples=5)
labels = dbscan.fit_predict(X_noisy)

# Step 3: Count clusters (excluding noise)
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
n_noise = np.sum(labels == -1)

print(f"Number of clusters found: {n_clusters}")
print(f"Number of noise points detected: {n_noise}")
```

Q.48. Load the Digits dataset, reduce dimensions using t-SNE, then apply Agglomerative Clustering and plot the
clusters.

Answer ->>

```python
# Import required libraries
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Step 1: Load the Digits dataset
digits = load_digits()
X = digits.data

# Step 2: Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 3: Reduce dimensions using t-SNE (to 2D)
tsne = TSNE(n_components=2, random_state=42, perplexity=30, learning_rate=200)
X_tsne = tsne.fit_transform(X_scaled)

# Step 4: Apply Agglomerative Clustering
agg = AgglomerativeClustering(n_clusters=10, linkage='ward')
labels = agg.fit_predict(X_tsne)

# Step 5: Visualize clusters
plt.figure(figsize=(8, 6))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=labels, cmap='tab10', s=40)
plt.title("Agglomerative Clustering on Digits Dataset (t-SNE 2D Projection)")
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")
plt.show()
```
